```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs #generate random values
from sklearn.cluster import KMeans
```

```python
#generate a sample data
sample_data=make_blobs(n_samples=100,n_features=2,centers=5,cluster_std=1.8,random_state=101)
```

```python
#plot the original data
plt.scatter(sample_data[0][:,0],sample_data[0][0:,1],c=sample_data[1],cmap='rainbow')
```

```python
#cluster using kmeans
kmeans=KMeans(n_clusters=5)
#train the algorithm with data
kmeans.fit(sample_data[0])
#get the cluster centers
kmeans.cluster_centers_
#cluster labels
kmeans.labels_
```

```python
#plot the graph
plt.scatter(sample_data[0][:,0],sample_data[0][:,1],c=kmeans.labels_,cmap='rainbow')
```
